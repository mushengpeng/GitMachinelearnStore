# 高级可视化seaborn 

任务要求 由于数据集拥有多项数据类型，通过成对数据的对比绘制出可视化图表，此处建议使用pairplot函数

```python
sns.pairplot(iris, hue='class', size=3,aspect=1);
```

任务提示 绘制多组直方图，对数据集的大致内容进行总体了解。此处使用hist函数绘制

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12,8));
plt.show();
```

任务要求  对数据的分布情况进行绘制，采用boxplot箱型图

```python
iris.boxplot(by='species', figsize=(14,14));
```

任务要求 需要对不同分类的数据的每一个特征进行了解，同时绘制出数据分布情况

此处涉及两个绘制重点

A 使用subplot在一个图中绘制多个数据 

B 使用seaborn的violinplot绘制超炫酷的小提琴图

```python
plt.figure(figsize=(12,8));
plt.subplot(2,2,1)
sns.violinplot(x='class', y='sepal length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='class', y='sepal width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='class', y='petal length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='class', y='petal width', data=iris);
```
